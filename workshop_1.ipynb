* **`~` (not)ï¼š** å°‡æ¢ä»¶åè½‰ (çœŸè®Šå‡ï¼Œå‡è®ŠçœŸ)ã€‚

* **ç¯©é¸é ­ç­‰è‰™ (Pclass = 1) çš„å¥³æ€§ä¹˜å®¢ï¼š** `df[(df['Pclass'] == 1) & (df['Sex'] == 'female')]`

## 8.1 åˆ†çµ„èšåˆ (GroupBy) çš„æ¦‚å¿µï¼šSplit-Apply-Combine

**`GroupBy`** æ˜¯ Pandas ä¸­**æœ€ç‚ºå¼·å¤§å’Œéˆæ´»**çš„è³‡æ–™åˆ†æå·¥å…·ä¹‹ä¸€ã€‚å®ƒå…è¨±ä½ æ ¹æ“šä¸€å€‹æˆ–å¤šå€‹æ¬„ä½çš„å€¼ï¼Œå°‡è³‡æ–™åˆ†æˆä¸åŒçš„çµ„ï¼Œç„¶å¾Œå°æ¯å€‹çµ„ç¨ç«‹åœ°åŸ·è¡Œæ“ä½œã€‚

1.  **Split (åˆ†å‰²)ï¼š** æ ¹æ“šä½ æŒ‡å®šçš„éµ (ä¸€å€‹æˆ–å¤šå€‹æ¬„ä½)ï¼Œå°‡ DataFrame åˆ†å‰²æˆå¤šå€‹è¼ƒå°çš„çµ„ (Group)ã€‚ä¾‹å¦‚ï¼Œä½ å¯ä»¥æŒ‰ `Sex` (æ€§åˆ¥) å°‡ä¹˜å®¢åˆ†æˆã€Œç”·æ€§çµ„ã€å’Œã€Œå¥³æ€§çµ„ã€ã€‚
2.  **Apply (æ‡‰ç”¨)ï¼š** å°æ¯å€‹ç¨ç«‹çš„çµ„ï¼Œæ‡‰ç”¨ä¸€å€‹å‡½æ•¸ã€‚é€™å¯ä»¥æ˜¯èšåˆå‡½æ•¸ (å¦‚ `mean()` å¹³å‡å€¼ã€`sum()` ç¸½å’Œã€`count()` è¨ˆæ•¸)ã€è½‰æ›å‡½æ•¸ã€æˆ–éæ¿¾å‡½æ•¸ã€‚ä¾‹å¦‚ï¼Œä½ å¯ä»¥è¨ˆç®—æ¯å€‹æ€§åˆ¥çµ„çš„ã€Œå¹³å‡ç”Ÿå­˜ç‡ã€ã€‚
3.  **Combine (çµ„åˆ)ï¼š** å°‡æ¯å€‹çµ„çš„è™•ç†çµæœçµ„åˆèµ·ä¾†ï¼Œå½¢æˆä¸€å€‹æ–°çš„ Series æˆ– DataFrameï¼Œä½œç‚ºæœ€çµ‚çš„è¼¸å‡ºã€‚

è®“æˆ‘å€‘ä¾†çœ‹çœ‹ä¸åŒæ€§åˆ¥çš„ç”Ÿå­˜ç‡æ˜¯å¤šå°‘ï¼Ÿä»¥åŠä¸åŒè‰™ç­‰çš„å¹³å‡ç¥¨åƒ¹ï¼Ÿ

````python
# ç¯„ä¾‹ï¼šè¨ˆç®—ä¸åŒæ€§åˆ¥çš„å¹³å‡ç”Ÿå­˜ç‡
# æ­¥é©Ÿï¼š
# 1. df.groupby('Sex')ï¼šå°‡è³‡æ–™æŒ‰ 'Sex' æ¬„ä½åˆ†æˆ 'female' å’Œ 'male' å…©çµ„
# 2. ['Survived']ï¼šå¾åˆ†çµ„å¾Œçš„ç‰©ä»¶ä¸­ï¼Œé¸å– 'Survived' æ¬„ä½
# 3. .mean()ï¼šå°æ¯ä¸€çµ„çš„ 'Survived' å€¼è¨ˆç®—å¹³å‡å€¼
survival_rate_by_sex = df.groupby('Sex')['Survived'].mean()

print("ä¸åŒæ€§åˆ¥çš„ç”Ÿå­˜ç‡ï¼š")
print(survival_rate_by_sex)
````

### ä½ çš„ä»»å‹™ç›®æ¨™

å˜—è©¦å°‡è³‡æ–™æ¸…ç†æˆä¸€å€‹ **æ›´é©åˆé€²è¡Œæ©Ÿå™¨å­¸ç¿’æ¨¡å‹è¨“ç·´** çš„ç‹€æ…‹ã€‚æ©Ÿå™¨å­¸ç¿’æ¨¡å‹é€šå¸¸åªèƒ½ç†è§£æ•¸å€¼ï¼Œæ‰€ä»¥æˆ‘å€‘çš„ç›®æ¨™æ˜¯ï¼š
1.  **è™•ç†æ‰€æœ‰ç¼ºå¤±å€¼**ã€‚
2.  **å°‡æ‰€æœ‰éæ•¸å€¼æ¬„ä½è½‰æ›ç‚ºæ•¸å€¼**ã€‚
3.  **ç§»é™¤ä¸éœ€è¦çš„æ¬„ä½**ã€‚
4.  **å‰µé€ å¯èƒ½æœ‰ç”¨çš„æ–°ç‰¹å¾µ (ç‰¹å¾µå·¥ç¨‹)**ã€‚

### å»ºè­°æ­¥é©Ÿ

* **è™•ç†ç¼ºå¤±å€¼ï¼š**
  * `Age`: ç”¨å¹³å‡å€¼æˆ–ä¸­ä½æ•¸å¡«å……ã€‚
  * `Embarked`: ç”¨çœ¾æ•¸å¡«å……ã€‚
  * `Cabin`: ç¼ºå¤±å€¼å¤ªå¤šï¼Œå¯ä»¥ç›´æ¥**åˆªé™¤**æ­¤æ¬„ä½ã€‚
* **è™•ç†ä¸éœ€è¦çš„æ¬„ä½ï¼š** `Name`, `Ticket`, `PassengerId` ç­‰æ¬„ä½å°é æ¸¬ç”Ÿå­˜ç‡æ²’æœ‰ç›´æ¥å¹«åŠ©ï¼Œå¯ä»¥è€ƒæ…®åˆªé™¤å®ƒå€‘ã€‚
* **å°‡é¡åˆ¥å‹è³‡æ–™è½‰æ›ç‚ºæ•¸å€¼å‹è³‡æ–™ï¼š**
  * `Sex` æ¬„ä½ ('female', 'male') å¯ä»¥è½‰æ›ç‚º `1` å’Œ `0`ã€‚
  * `Embarked` æ¬„ä½ ('S', 'C', 'Q')ï¼Œå› ç‚ºæœ‰å¤šå€‹é¡åˆ¥ï¼Œéœ€è¦ä½¿ç”¨ **ç¨ç†±ç·¨ç¢¼ (One-Hot Encoding)** ä¾†è™•ç†ã€‚Pandas çš„ `pd.get_dummies()` å‡½æ•¸éå¸¸é©åˆé€™å€‹ä»»å‹™ã€‚
* **å‰µå»ºæ–°çš„ç‰¹å¾µ (Feature Engineering)ï¼š** ä½ å¯ä»¥åƒæˆ‘å€‘ä¹‹å‰æ–°å¢ `FamilySize` æ¬„ä½é‚£æ¨£ï¼Œå¾ç¾æœ‰æ•¸æ“šä¸­çµ„åˆå‡ºæ–°çš„æœ‰æ„ç¾©çš„æ¬„ä½ã€‚

### è‡ªç”±ç™¼æ®ï¼

è«‹ç›¡é‡ç™¼æ®ä½ çš„å‰µæ„ï¼Œæ²’æœ‰æ¨™æº–ç­”æ¡ˆï¼æˆ‘å€‘é¼“å‹µä½ å˜—è©¦ä¸åŒçš„æ–¹æ³•ã€‚åœ¨ä½ å®Œæˆæ¸…ç†å¾Œï¼Œè«‹å‘½åä½ çš„æœ€çµ‚ DataFrame ç‚º `df_cleaned_for_dl`ã€‚

---

````python
# åœ¨æ­¤é€²è¡Œä½ çš„è³‡æ–™æ¸…ç†èˆ‡é è™•ç†ï¼

# å¼·çƒˆå»ºè­°ï¼šé¦–å…ˆè¤‡è£½åŸå§‹è³‡æ–™ï¼Œé¿å…ä¿®æ”¹åˆ°åŸå§‹çš„ df
df_cleaned_for_dl = df.copy()

# --- ä»¥ä¸‹æ˜¯ä½ å¯ä»¥é–‹å§‹å¯«ç¨‹å¼ç¢¼çš„åœ°æ–¹ (ä½ å¯ä»¥æ ¹æ“šä½ çš„æƒ³æ³•åˆªé™¤æˆ–ä¿®æ”¹é€™äº›ç¯„ä¾‹ç¨‹å¼ç¢¼) --- 

print("æ¸…ç†å‰çš„ç¼ºå¤±å€¼ï¼š\n", df_cleaned_for_dl.isnull().sum())
print("\n--- é–‹å§‹æ¸…ç† ---\n")

# 1. è™•ç†ç¼ºå¤±å€¼
# è™•ç† 'Age' ç¼ºå¤±å€¼ï¼šç”¨å¹³å‡å€¼å¡«å……
df_cleaned_for_dl['Age'].fillna(df_cleaned_for_dl['Age'].mean(), inplace=True)

# è™•ç† 'Embarked' ç¼ºå¤±å€¼ï¼šç”¨çœ¾æ•¸å¡«å……
df_cleaned_for_dl['Embarked'].fillna(df_cleaned_for_dl['Embarked'].mode()[0], inplace=True)

# 2. åˆªé™¤ä½ èªç‚ºä¸éœ€è¦çš„æ¬„ä½
# 'Cabin' ç¼ºå¤±å€¼å¤ªå¤šï¼Œ'Name', 'Ticket', 'PassengerId' å°æ¨¡å‹ä¾†èªªæ˜¯é›œè¨Š
df_cleaned_for_dl.drop(['Cabin', 'Name', 'Ticket', 'PassengerId'], axis=1, inplace=True)

# 3. å°‡é¡åˆ¥å‹è³‡æ–™è½‰æ›ç‚ºæ•¸å€¼
# å°‡ 'Sex' æ¬„ä½è½‰æ›ç‚ºæ•¸å€¼ (0 æˆ– 1)
# .map() æ˜¯ä¸€å€‹æ–¹ä¾¿çš„è½‰æ›å·¥å…·
df_cleaned_for_dl['Sex'] = df_cleaned_for_dl['Sex'].map({'female': 1, 'male': 0})

# å° 'Embarked' é€²è¡Œç¨ç†±ç·¨ç¢¼ (One-Hot Encoding)
# drop_first=True å¯ä»¥é¿å…ã€Œå…±ç·šæ€§ã€å•é¡Œï¼Œé€™æ˜¯æ©Ÿå™¨å­¸ç¿’ä¸­çš„ä¸€å€‹å¥½ç¿’æ…£
df_cleaned_for_dl = pd.get_dummies(df_cleaned_for_dl, columns=['Embarked'], drop_first=True)

# 4. æ–°å¢ 'FamilySize' æ¬„ä½ (ç‰¹å¾µå·¥ç¨‹)
df_cleaned_for_dl['FamilySize'] = df_cleaned_for_dl['SibSp'] + df_cleaned_for_dl['Parch'] + 1

# --- ä½ çš„æ¸…ç†å·¥ä½œåˆ°æ­¤çµæŸï¼Œè«‹ç¢ºä¿æœ€çµ‚ DataFrame åç¨±ç‚º df_cleaned_for_dl ---

print("\n--- æ¸…ç†å®Œæˆ ---\n")
print("ä½ çš„æ¸…ç†æˆæœï¼š")
print("æ¸…ç†å¾Œçš„è³‡æ–™å½¢ç‹€ï¼š", df_cleaned_for_dl.shape)
print("\næ¸…ç†å¾Œçš„ç¼ºå¤±å€¼ï¼š\n", df_cleaned_for_dl.isnull().sum()) # æ‡‰è©²éƒ½æ²’æœ‰ç¼ºå¤±å€¼äº†
print("\næ¸…ç†å¾Œè³‡æ–™çš„å‰ 5 è¡Œï¼š")
print(df_cleaned_for_dl.head())
print("\næ¸…ç†å¾Œè³‡æ–™çš„æ‰€æœ‰æ¬„ä½ï¼š", df_cleaned_for_dl.columns.tolist())
````

## 11.1 æ·±åº¦å­¸ç¿’å°æ¨¡å‹ï¼šçœ‹çœ‹ä½ çš„è³‡æ–™å½±éŸ¿åŠ›ï¼

ç¾åœ¨ï¼Œè®“æˆ‘å€‘çœ‹çœ‹ä½ æ¸…ç†å¥½çš„è³‡æ–™å¦‚ä½•å½±éŸ¿ä¸€å€‹ç°¡å–®çš„**æ·±åº¦å­¸ç¿’æ¨¡å‹**ï¼

æˆ‘å€‘å°‡ä½¿ç”¨ **PyTorch**ï¼Œä¸€å€‹å¼·å¤§çš„é–‹æºæ©Ÿå™¨å­¸ç¿’æ¡†æ¶ï¼Œä¾†å»ºç«‹ä¸€å€‹éå¸¸ç°¡å–®çš„ç¥ç¶“ç¶²è·¯ï¼Œé æ¸¬ä¹˜å®¢çš„ç”Ÿå­˜ç‡ã€‚ä½ **ä¸éœ€è¦ç†è§£é€™å€‹æ¨¡å‹çš„æ¯ä¸€å€‹ç´°ç¯€**ï¼Œé€™å€‹ç’°ç¯€çš„é‡é»æ˜¯è§€å¯Ÿï¼š

* **è³‡æ–™è™•ç†æ˜¯æ©Ÿå™¨å­¸ç¿’æµç¨‹ä¸­ä¸å¯æˆ–ç¼ºçš„ä¸€ç’°ã€‚**
* **å¥½çš„è³‡æ–™å“è³ªå°æ¨¡å‹è¡¨ç¾è‡³é—œé‡è¦ (Garbage In, Garbage Out)ã€‚**
* é€™æ˜¯ä½ ç‚ºæœªä¾†æ·±åº¦å­¸ç¿’å·¥ä½œåŠå»ºç«‹é€£çµçš„**ç¬¬ä¸€å€‹çœŸå¯¦é«”é©—**ï¼

è«‹é‹è¡Œä»¥ä¸‹çš„ Cellsï¼Œçœ‹çœ‹ä½ çš„è³‡æ–™æ¸…ç†æˆæœèƒ½è®“æ¨¡å‹é”åˆ°å¤šå°‘çš„é æ¸¬æº–ç¢ºåº¦ (Accuracy)ï¼

---

````python
# å°å…¥ PyTorch å’Œå…¶ä»–å¿…è¦çš„å‡½å¼åº«
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

print("PyTorch å’Œç›¸é—œå‡½å¼åº«è¼‰å…¥æˆåŠŸï¼")

# --- è³‡æ–™æº–å‚™ï¼šå°‡ DataFrame è½‰æ›ç‚ºæ¨¡å‹å¯ç”¨çš„æ ¼å¼ ---

# å†æ¬¡æª¢æŸ¥ä½ çš„ df_cleaned_for_dlï¼Œç¢ºä¿å®ƒæ˜¯å­˜åœ¨çš„
try:
    df_cleaned_for_dl
except NameError:
    print("éŒ¯èª¤ï¼šæ‰¾ä¸åˆ° df_cleaned_for_dlï¼Œè«‹ç¢ºèªä½ å·²ç¶“é‹è¡Œäº†ä¸Šé¢çš„é–‹æ”¾å¼ç·´ç¿’ Cellï¼")
    # å¦‚æœå­¸å“¡æ²’æœ‰é‹è¡Œï¼Œé€™è£¡å¯ä»¥é¸æ“‡åœæ­¢æˆ–æä¾›ä¸€å€‹é è¨­çš„æ¸…ç†ç‰ˆæœ¬
    # ç‚ºäº†å·¥ä½œåŠé †åˆ©ï¼Œæˆ‘å€‘å‡è¨­å®ƒå­˜åœ¨

# é¸æ“‡ç”¨æ–¼æ¨¡å‹è¨“ç·´çš„ç‰¹å¾µ (è¼¸å…¥è®Šæ•¸ X) å’Œç›®æ¨™è®Šæ•¸ (é æ¸¬ç›®æ¨™ y)
# ç›®æ¨™æ˜¯ 'Survived'ï¼Œå…¶ä»–æ‰€æœ‰æ¬„ä½éƒ½æ˜¯ç‰¹å¾µ
X = df_cleaned_for_dl.drop('Survived', axis=1)
y = df_cleaned_for_dl['Survived']

# è³‡æ–™æ¨™æº–åŒ– (Standardization)ï¼š
# é€™æ˜¯æ·±åº¦å­¸ç¿’ä¸­éå¸¸é‡è¦çš„ä¸€æ­¥ã€‚å› ç‚ºä¸åŒç‰¹å¾µçš„æ•¸å€¼ç¯„åœå·®ç•°å¾ˆå¤§ (å¦‚ Age å’Œ Fare)ï¼Œ
# æ¨™æº–åŒ–å¯ä»¥å°‡æ‰€æœ‰ç‰¹å¾µç¸®æ”¾åˆ°ç›¸ä¼¼çš„ç¯„åœï¼Œæœ‰åŠ©æ–¼æ¨¡å‹æ›´å¿«ã€æ›´ç©©å®šåœ°å­¸ç¿’ã€‚
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# åˆ†å‰²è¨“ç·´é›†å’Œæ¸¬è©¦é›†ï¼š
# æˆ‘å€‘å°‡è³‡æ–™åˆ†ç‚ºã€Œè¨“ç·´é›†ã€(ç”¨ä¾†è¨“ç·´æ¨¡å‹)å’Œã€Œæ¸¬è©¦é›†ã€(ç”¨ä¾†è©•ä¼°æ¨¡å‹è¡¨ç¾)ã€‚
# test_size=0.2 è¡¨ç¤º 20% çš„è³‡æ–™ç”¨æ–¼æ¸¬è©¦ã€‚
# random_state=42 ç¢ºä¿æ¯æ¬¡åˆ†å‰²çµæœéƒ½ä¸€æ¨£ï¼Œæ–¹ä¾¿é‡ç¾å¯¦é©—ã€‚
# [ä¿®æ”¹] é€™è£¡æˆ‘å€‘å‚³å…¥ y (Pandas Series)ï¼Œè€Œä¸æ˜¯ y.valuesï¼Œé€™æ¨£ y_test å°±æœƒä¿ç•™åŸå§‹çš„ç´¢å¼•ï¼Œæ–¹ä¾¿å¾ŒçºŒåˆ†æ
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# è½‰æ›ç‚º PyTorch å¼µé‡ (Tensor)ï¼šPyTorch æ¨¡å‹éœ€è¦è¼¸å…¥å¼µé‡æ ¼å¼çš„æ•¸æ“š
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).unsqueeze(1) # unsqueeze(1) å°‡ä¸€ç¶­å¼µé‡è®Šç‚ºäºŒç¶­åˆ—å‘é‡
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).unsqueeze(1)

print("\nè³‡æ–™æº–å‚™å®Œæˆï¼")
print(f"ç”¨æ–¼è¨“ç·´çš„ç‰¹å¾µæ¬„ä½ ({X.shape[1]}å€‹): {X.columns.tolist()}")
print("è¨“ç·´é›†ç‰¹å¾µå½¢ç‹€ï¼š", X_train_tensor.shape)
print("æ¸¬è©¦é›†ç‰¹å¾µå½¢ã‚‰ãªã‹ã£ãŸï¼š", X_test_tensor.shape)
````

````python
# [ä¿®æ”¹] å»ºç«‹ä¸¦è¨“ç·´ä¸€å€‹ç°¡å–®çš„ PyTorch ç¥ç¶“ç¶²è·¯ (å¢åŠ  Loss è¨˜éŒ„)

# 1. å®šç¾©ç¥ç¶“ç¶²è·¯æ¨¡å‹
class SimpleNN(nn.Module):
    def __init__(self, input_size):
        super(SimpleNN, self).__init__()
        self.layer1 = nn.Linear(input_size, 64)
        self.relu = nn.ReLU()
        self.layer2 = nn.Linear(64, 32)
        self.relu2 = nn.ReLU()
        self.output_layer = nn.Linear(32, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.layer1(x)
        x = self.relu(x)
        x = self.layer2(x)
        x = self.relu2(x)
        x = self.output_layer(x)
        x = self.sigmoid(x)
        return x

# 2. å¯¦ä¾‹åŒ–æ¨¡å‹ã€å®šç¾©æå¤±å‡½æ•¸å’Œå„ªåŒ–å™¨
input_size = X_train_tensor.shape[1]
model = SimpleNN(input_size)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 3. è¨“ç·´æ¨¡å‹
num_epochs = 100
losses = [] # [æ–°å¢] ç”¨ä¾†è¨˜éŒ„æ¯ä¸€è¼ªçš„ loss
print(f"\né–‹å§‹è¨“ç·´æ·±åº¦å­¸ç¿’æ¨¡å‹ï¼Œå…± {num_epochs} å€‹ Epoch (è¨“ç·´è¼ªæ•¸)..." )

for epoch in range(num_epochs):
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    losses.append(loss.item()) # [æ–°å¢] è¨˜éŒ„ loss
    
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

print("\næ¨¡å‹è¨“ç·´å®Œæˆï¼")
````

````python
# è©•ä¼°æ¨¡å‹è¡¨ç¾ï¼šé æ¸¬ç”Ÿå­˜ç‡

print("\né–‹å§‹è©•ä¼°æ¨¡å‹åœ¨ã€æ¸¬è©¦é›†ã€ä¸Šçš„è¡¨ç¾...")

# è¨­ç½®æ¨¡å‹ç‚ºè©•ä¼°æ¨¡å¼
model.eval()

# åœ¨æ¸¬è©¦é›†ä¸Šé€²è¡Œé æ¸¬
with torch.no_grad(): 
    test_outputs = model(X_test_tensor)
    predicted = (test_outputs >= 0.5).float()
    accuracy = accuracy_score(y_test_tensor.numpy(), predicted.numpy())

print(f"\næ¨¡å‹åœ¨æ¸¬è©¦é›†ä¸Šçš„æº–ç¢ºåº¦ (Accuracy): {accuracy:.4f} ({accuracy*100:.2f} %)")

print("\n---")
print("æ€è€ƒï¼šä½ çš„è³‡æ–™æ¸…ç†æ–¹æ³•æ˜¯å¦è®“é€™å€‹æº–ç¢ºåº¦æ›´é«˜æˆ–æ›´ä½ï¼Ÿ")
print("å˜—è©¦å›åˆ° \"10.1 é–‹æ”¾å¼ç·´ç¿’\" å€å¡Šï¼Œä¿®æ”¹ä½ çš„æ¸…ç†æ–¹å¼ï¼Œç„¶å¾Œé‡æ–°é‹è¡Œæ‰€æœ‰æ·±åº¦å­¸ç¿’çš„ Cellsï¼Œè§€å¯Ÿæº–ç¢ºåº¦çš„è®ŠåŒ–ï¼")
print("---")
````

### 12.1 è¦–è¦ºåŒ–å­¸ç¿’éç¨‹ï¼šæ¨¡å‹æ˜¯å¦‚ä½•é€²æ­¥çš„ï¼Ÿ

````python
# ç¹ªè£½ Loss æ›²ç·š
plt.figure(figsize=(10, 5))
plt.plot(losses)
plt.title('Training Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
````

### 12.2 åˆ†ææ¨¡å‹çš„éŒ¯èª¤ï¼šå®ƒåœ¨å“ªäº›äººèº«ä¸Šçœ‹èµ°äº†çœ¼ï¼Ÿ

æ¨¡å‹ä¸æ˜¯è¬èƒ½çš„ï¼Œå®ƒç¸½æœƒçŠ¯éŒ¯ã€‚æˆ‘å€‘çš„ç›®æ¨™æ˜¯æ‰¾å‡ºè¢«æ¨¡å‹ã€Œèª¤åˆ¤ã€çš„ä¹˜å®¢ï¼Œçœ‹çœ‹ä»–å€‘æœ‰æ²’æœ‰ä»€éº¼å…±åŒç‰¹å¾µã€‚é€™èƒ½å¹«åŠ©æˆ‘å€‘ç†è§£æ¨¡å‹çš„å¼±é»ï¼Œç”šè‡³å•Ÿç™¼æˆ‘å€‘å›åˆ°å‰é¢çš„æ­¥é©Ÿï¼Œåšæ›´å¥½çš„ã€Œç‰¹å¾µå·¥ç¨‹ã€ã€‚

æˆ‘å€‘æœƒçœ‹å…©ç¨®ä¸»è¦çš„éŒ¯èª¤ï¼š
* **å½é™°æ€§ (False Negative)**ï¼šä¹˜å®¢æ˜æ˜**å­˜æ´»**äº†ï¼Œæ¨¡å‹å»é æ¸¬ä»–**æ­»äº¡**ã€‚(æ¨¡å‹éæ–¼æ‚²è§€)
* **å½é™½æ€§ (False Positive)**ï¼šä¹˜å®¢æ˜æ˜**æ­»äº¡**äº†ï¼Œæ¨¡å‹å»é æ¸¬ä»–**å­˜æ´»**ã€‚(æ¨¡å‹éæ–¼æ¨‚è§€)

````python
# æ­¥é©Ÿ 1: æ‰¾å‡ºæ‰€æœ‰éŒ¯èª¤æ¡ˆä¾‹

# å°‡ PyTorch Tensor è½‰å› Numpy arrayï¼Œæ–¹ä¾¿æ¯”è¼ƒ
y_test_numpy = y_test.values
predicted_numpy = predicted.numpy().flatten() # .flatten() å°‡ [[0],[1],...] è½‰ç‚º [0,1,...]

# æ‰¾å‡ºé æ¸¬èˆ‡å¯¦éš›ä¸ç¬¦çš„æ¨£æœ¬
error_mask = (y_test_numpy != predicted_numpy)
error_indices = y_test[error_mask].index

# å¾åŸå§‹çš„ df ä¸­ï¼Œåˆ©ç”¨ç´¢å¼•æ‰¾å‡ºé€™äº›éŒ¯èª¤æ¡ˆä¾‹çš„å®Œæ•´è³‡æ–™
df_errors = df.loc[error_indices].copy() # .copy() æ˜¯å€‹å¥½ç¿’æ…£

# æ–°å¢æ¬„ä½ï¼Œæ–¹ä¾¿æˆ‘å€‘å°ç…§ã€ŒçœŸå¯¦ã€èˆ‡ã€Œé æ¸¬ã€çµæœ
df_errors['Actual_Survived'] = y_test[error_mask]
df_errors['Predicted_Survived'] = predicted_numpy[error_mask].astype(int)

print(f"åœ¨ {len(y_test)} å€‹æ¸¬è©¦æ¨£æœ¬ä¸­ï¼Œæ¨¡å‹ç¸½å…±é æ¸¬éŒ¯èª¤äº† {len(df_errors)} å€‹ã€‚")
print("\nä»¥ä¸‹æ˜¯éƒ¨åˆ†éŒ¯èª¤æ¡ˆä¾‹ï¼š")
df_errors.head()
````

### 12.3 éŒ¯èª¤æ¡ˆä¾‹åˆ†ä½ˆ vs. æ•´é«”åˆ†ä½ˆ

> **è¢«æ¨¡å‹èª¤åˆ¤çš„é€™ç¾¤äººï¼Œä»–å€‘çš„åˆ†ä½ˆï¼Œå’Œå…¨éƒ¨çš„æ¸¬è©¦ä¹˜å®¢åˆ†ä½ˆï¼Œæœ‰ä»€éº¼ä¸ä¸€æ¨£å—ï¼Ÿ**

å¦‚æœæˆ‘å€‘ç™¼ç¾ï¼ŒéŒ¯èª¤æ¡ˆä¾‹ä¸­ã€Œä¸‰ç­‰è‰™ã€çš„æ¯”ä¾‹ç‰¹åˆ¥é«˜ï¼Œé€™å¯èƒ½å°±ä»£è¡¨æ¨¡å‹å°æ–¼åˆ¤æ–·ä¸‰ç­‰è‰™ä¹˜å®¢çš„ç”Ÿæ­»æ¯”è¼ƒæ²’æŠŠæ¡ã€‚è®“æˆ‘å€‘ä¾†æ¯”è¼ƒçœ‹çœ‹ `Pclass` (è‰™ç­‰) å’Œ `Sex` (æ€§åˆ¥) çš„åˆ†ä½ˆã€‚

````python
# æ­¥é©Ÿ 2: æ¯”è¼ƒéŒ¯èª¤æ¡ˆä¾‹èˆ‡æ•´é«”æ¸¬è©¦é›†çš„åˆ†ä½ˆ

# å…ˆå–å¾—æ•´å€‹æ¸¬è©¦é›†çš„åŸå§‹è³‡æ–™ï¼Œæ–¹ä¾¿å°æ¯”
df_test_set = df.loc[y_test.index]

print("--- Pclass (è‰™ç­‰) åˆ†ä½ˆæ¯”è¼ƒ ---")
print("\næ•´é«”æ¸¬è©¦é›† Pclass åˆ†ä½ˆ:")
print(df_test_set['Pclass'].value_counts(normalize=True).sort_index())
print("\néŒ¯èª¤æ¡ˆä¾‹ Pclass åˆ†ä½ˆ:")
print(df_errors['Pclass'].value_counts(normalize=True).sort_index())

print("\n--- Sex (æ€§åˆ¥) åˆ†ä½ˆæ¯”è¼ƒ ---")
print("\næ•´é«”æ¸¬è©¦é›† Sex åˆ†ä½ˆ:")
print(df_test_set['Sex'].value_counts(normalize=True).sort_index())
print("\néŒ¯èª¤æ¡ˆä¾‹ Sex åˆ†ä½ˆ:")
print(df_errors['Sex'].value_counts(normalize=True).sort_index())
````

### 12.4 å€‹æ¡ˆåˆ†æï¼šçœ‹çœ‹å¹¾å€‹å…·é«”çš„ä¾‹å­
